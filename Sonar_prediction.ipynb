```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv(r'c:\\Users\\mahen\\OneDrive\\Desktop\\sonar.csv', header =None)
```

```python
X=sonar_data.drop(columns=60,axis=1)
y=sonar_data[60]
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, random_state=1,test_size=0.1,stratify=y)
```

```python
X_train
```

```python
y_train
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model= LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Accuracy checking on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
```

```python
print('Accuracy on training data:', training_data_accuracy)
```

```python
#Accuracy checking on test data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,y_test)
```

```python
print('Accuracy on training data:', testing_data_accuracy)
```
