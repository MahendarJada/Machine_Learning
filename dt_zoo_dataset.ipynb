```python
import matplotlib.pyplot as plt
for col_name in df.columns:
    plt.figure()
    plt.hist(df[col_name])
    plt.title(col_name)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42, test_size=0.3)
```

```python
dt_clf=DecisionTreeClassifier(random_state=42)
dt_clf=dt_clf.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(15,15))
plot_tree(dt_clf, filled=True,rounded=True,class_names=['Mammal','Bird','Fish','Amphiban','Reptile','Bug','Invertebrate'],feature_names=X.columns)
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(dt_clf,X_test,y_test,display_labels=['Mammal','Bird','Fish','Amphiban','Reptile','Bug','Invertebrate'])
```

```python
predictions=dt_clf.predict(X_test)
```
