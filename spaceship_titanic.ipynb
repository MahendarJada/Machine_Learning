```python
X_train=train.drop(columns=['Transported'])
X_train
```

```python
y_train=train['Transported']
y_train
```

```python
X_test=pd.read_csv(r'c:\\Users\\mahen\\OneDrive\\Desktop\\datasets\\spaceship_titanic\\test.csv')
X_test
```

```python
test_index=X_test['PassengerId']
test_index
```

```python
X_train = X_train.drop(columns=["PassengerId", "Name"])
X_test = X_test.drop(columns=["PassengerId", "Name"])

X_train["Cabin"] = X_train["Cabin"].str[-1:]
X_test["Cabin"] = X_test["Cabin"].str[-1:]
print(X_train)
print(X_test)
```

```python
X_train['CryoSleep']=X_train['CryoSleep'].replace({True:1, False:-1})
X_train['CryoSleep']=X_train['CryoSleep'].fillna(0)
print(X_train)
```

```python
X_test['CryoSleep']=X_test['CryoSleep'].replace({True:1, False:-1})
X_test['CryoSleep']=X_test['CryoSleep'].fillna(0)
print(X_test)
```

```python
X_train=pd.get_dummies(X_train,columns=['HomePlanet','Cabin','Destination'])
X_test=pd.get_dummies(X_test,columns=['HomePlanet','Cabin','Destination'])
print(X_train)
print(X_test)
```

```python
X_train['VIP']*=1
X_train['VIP']=X_train['VIP'].fillna(0)
X_train['VIP']
```

```python
X_test['VIP']*=1
X_test['VIP']=X_test['VIP'].fillna(0)
X_test['VIP']
```

```python
X_train.head()
```

```python
from xgboost import XGBClassifier
```

```python
model=XGBClassifier(n_estimators=20, random_State=71)
model.fit(X_train,y_train)
```

```python
pred=model.predict_proba(X_test)[:,1]
```

```python
pred_label = np.where(pred > 0.5, True, False)
pred_label
```

```python
pred_label.shape
```

```python
submission = pd.DataFrame({"PassengerId": test_index, "Transported": pred_label})
submission
```
